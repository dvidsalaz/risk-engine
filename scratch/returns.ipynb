```python
#Expected Return of An Asset using Historical Returns

import numpy as np

r_values = [.01, -.005, .02, .004, -.01] #example/ arbitrary values | Rt

def expected_return_asset(returns):
    """
    Calculated the expected return of an asset as arithmetic mean of historical returns.
    Parameters: returns as historical returns

    Returns:
    float: expected_return

    """
    return np.mean(returns)

expected_return_asset(r_values)
```
