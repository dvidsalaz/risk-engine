```python
#Historical Average Return | Math Notation: Ri,t

#Ri,t: Arbitrary Sample Data -- will introduce real data later on
returns = [0.02, -0.0098, 0.0297]

def historical_avg_return(returns):

    sum_returns = 0

    for r in returns:
        sum_returns += r
    
    n_periods = len(returns)
    
    avg_return = sum_returns/n_periods

    return(avg_return)

historical_avg_return(returns)
```

```python
#Historical Sample Variance



def historical_variance(returns):

    sum_squared_deviations = 0

    avg_return = historical_avg_return(returns)

    for r in returns: #looping through the elements

        sum_squared_deviations += (r - avg_return) ** 2

    n_periods = len(returns)

    variance = sum_squared_deviations / (n_periods - 1)

    return(variance)

historical_variance(returns)
    




```

```python
#Standard Deviation of Variance == Volatility

import math

var = historical_variance(returns)

std_dev = math.sqrt(var)

print(std_dev)
```
